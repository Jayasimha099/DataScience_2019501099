***Handling big data and Data Scraping:***

**A.	find the 20 most popular movies with a rank more than 8.0**

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
basics = pd.read_csv("titlebasics.tsv", delimiter = "\t")
basics.head(5)
```

```python
ratings = pd.read_csv("titleratings.tsv", delimiter = "\t")
ratings.head(5)
```

```python
data = ratings[ratings['averageRating'] > 8.0]
data.head(5)
```

```python
final_ratings = data.sort_values('numVotes', ascending = False)
```

```python
final_ratings.head(20)
```

```python
output = pd.merge(final_ratings, basics, on= 'tconst')
output.head(10)
```

```python
output = output[output['titleType'] == 'movie']
```

```python
output.head(10)
```

```python
output['originalTitle'].head(20)
```

**B. find the 20 best rated movies with over 40,000 votes in the 2000s (year >= 2000)**

```python
basics.dtypes
```

```python
basics['startYear'] = pd.to_numeric(basics['startYear'], errors='coerce')
```

```python
basics.dtypes
```

```python
year = basics[(basics['startYear'] >= 2000) & (basics['titleType'] == 'movie')]
year.head(10)
```

```python
output1 = pd.merge(year, ratings, on= 'tconst')
output1.head(10)
```

```python
votes = output1[output1['numVotes'] > 40000]
votes.head(10)
```

```python
Output2 = votes.sort_values('averageRating', ascending = False)
Output2.head(20)
```

```python
Output2['originalTitle'].head(20)
```
